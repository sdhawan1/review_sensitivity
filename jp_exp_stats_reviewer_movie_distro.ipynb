```python
#get all the data (i.e. numbers corresponding to freq. of review of movie)
movierevs = [1]
movierevs_high = [6] #will store all movies with more than 6 reviews
lineno = 0
with open("revspermovie") as openfileobject:
    for line in openfileobject:
        lineno = int(line)
        movierevs.extend([lineno])
        if lineno > 5:
            movierevs_high.extend([lineno])
```

```python
#this outputs the total number of reviews; just a way to verify the frequencies found above.
sum = 0
movies = 0
for i in movierevs:
    sum += i
    movies += 1
print "sum: " + str(sum)
print "movies: " + str(movies)
```

```python
import matplotlib
import matplotlib.pyplot
%matplotlib inline

matplotlib.pyplot.hist(movierevs, 30)
matplotlib.pyplot.title("histogram of reviews per movie (all movies)")
```

```python
#get all the reviewer data (i.e. numbers corresponding to freq. of reviewer)
criticrevs = [1] #gets all the raw review numbers
criticrevs_high = [10] #gets only the reviewers with a large number of reviews
lineno = 0
with open("generated_data/revspercritic") as openfileobject:
    for line in openfileobject:
        lineno = int(line)
        criticrevs.extend([lineno])
        if lineno >= 10:
            criticrevs_high.extend([lineno])

#some statistics
highrevs = 0
for num in criticrevs_high:
    if (num < 100) and (num > 0):
        highrevs += num
print highrevs
```

```python
#this outputs the total number of reviews
sum = 0
critics = 0
for i in criticrevs:
    sum += i
    critics += 1
print "sum: " + str(sum)
print "reviewers: " + str(critics) 
```
