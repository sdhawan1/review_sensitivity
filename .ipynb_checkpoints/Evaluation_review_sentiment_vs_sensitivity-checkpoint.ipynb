```python
#DON'T NEED TO RUN THIS AGAIN.
#now, take all of these results and parse scores by movie. 
#create a file that contains, for all movies, for all authors, an index matched to positive and negative scores.

from collections import defaultdict
import os
resfiles = os.listdir("tempres/")

#for each result, create an index matching a movie index to its list of reviewer indices.
movs = defaultdict(list)
for f in resfiles:
    movind = int(f.split('-')[1])
    revind = int(f.split('-')[2])
    movs[movind] += [revind]

#for every key in the dictionary (i.e. every movie), create a file with the reviewers and their scores
for key in movs.keys():
    wtxt = ""
    for rev in movs[key]:
        fres = open("tempres/res-"+str(key)+"-"+str(rev))
        ftxtarr = fres.read().split("---------------------------------------------------------------------------")
        posnegline = ftxtarr[1].split("\n")
        wds = posnegline[1].split("=")
        pos = float(wds[1][:-4])
        neg = float(wds[2])
        wtxt += str(rev) + " " + str(pos) + " " + str(neg) + "\n"
    #write the overall results into a file
    fout = open("generated_data/movie_sentiments/res_"+str(key)+"_mod", "w")
    fout.write(wtxt)
    fout.close()


```
