```python
%%bash
for f in ./temp/pos/*
do
    echo $f
    python ~/sentiment_classifier-0.5/src/senti_classifier/senti_classifier.py >>res_pos -c $f
done

for f in ./temp/neg/*
do
    echo $f
    python ~/sentiment_classifier-0.5/src/senti_classifier/senti_classifier.py >>res_neg -c $f
done

#this cell: should evaluate a large number (~50 to 100) of positive reviews and record the results
#goal is to test the sentiment classifier
```

```python
## open the file with the positive document and compute the total number of positive scores
import re

f =  open("res_pos")
ftxt = f.read()
frevs = ftxt.split('---------------------------------------------------------------------------')
#print frevs[1:4]

fscores = frevs[1::2]
fscorelines_pos = []
line = ""
positives = 0
for score in fscores:
    line = score.split('\n')[1]
    if re.search('Positive', line):
        print line
        positives += 1
    fscorelines_pos += [line]
    
print positives
```

```python
## open the file with negative scores and compute the percentage of negative scores.

f =  open("res_neg")
ftxt = f.read()
frevs = ftxt.split('---------------------------------------------------------------------------')
#print frevs[1:4]

fscores = frevs[1::2]
fscorelines_neg = []
negatives = 0
for score in fscores:
    line = score.split('\n')[1]
    if re.search('Negative', line):
        print score.split('\n')[1]
        negatives += 1
    fscorelines_neg += [line]
print negatives
```

```python
import numpy
posratios = []
for line in fscorelines_pos:
    linearr = line.split('=')
    pos = float(linearr[-2][:-4])
    neg = float(linearr[-1])
    posratios += [pos/(neg+pos)]

print sum(posratios)/len(posratios)
print numpy.std(posratios)
```
