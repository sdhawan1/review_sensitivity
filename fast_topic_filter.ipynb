```python
## create a faster algorithm to filter the words
#BEST TOPIC WORDS ARE BEFORE LINE 237 IN FREQ_TOPICS_WORD_MOD

#here, simply create a freqdist for the entire brown corpus, filter out the stopwords and the non-isalpha words.

import nltk
import sys
from nltk.probability import FreqDist
from nltk.corpus import brown
from nltk.corpus import stopwords

#This creates the frequency distribution much more quickly.
fd_brown = FreqDist(brown.words())

#now, remove the stopwords
for wd in stopwords.words("english"):
    if fd_brown[wd] > 0: #if the word is actually inside brown corpus
        del fd_brown[wd]

i = 0
for wd in fd_brown.keys():
    if not wd.isalpha():
        del fd_brown[wd]
```

```python
# now get all tokens from about 2000 movie reviews.
import os
path = os.path.abspath(os.path.join(os.getcwd(),".."))
negpath = path + "/data/text_only_reviews/txt_sentoken/neg/"
pospath = path + "/data/text_only_reviews/txt_sentoken/pos/"

alltokens = []
negfilenames = os.listdir(negpath)
posfilenames = os.listdir(pospath)
i = 0
for rev in negfilenames:
    i += 1
    if (i%50 == 0):
        sys.stdout.write(".")
    f = open(negpath + rev, "r")
    alltokens = alltokens + nltk.word_tokenize(f.read())

print "finished negs, starting pos..."

for rev in posfilenames:
    i += 1
    if (i%50 == 0):
        sys.stdout.write(".")
    f = open(pospath + rev, "r")
    alltokens = alltokens + nltk.word_tokenize(f.read())

#demo - getting tokens from a file quickly:
"""
f = open("reviewerdata/reviewers1", "r")
ftxt = f.read()
ftokens = nltk.word_tokenize(ftxt)
ffd = FreqDist(ftokens) #and kablamo, now you have a FreqDist of the tokens in f.
"""
```

```python
#create freqdist of all tokens of movie reviews, and then filter out stopwords and non-alphabetical tokens

fd_movie = FreqDist(alltokens)

#now, remove the stopwords
for wd in stopwords.words("english"):
    if fd_movie[wd] > 0: #if the word is actually inside brown corpus
        del fd_movie[wd]

i = 0
for wd in fd_movie.keys():
    if not wd.isalpha():
        del fd_movie[wd]
```

```python
#now, perform a larger bayesian analysis to determine TOPICALITY, like in "initial word filtering"
freqratio = {}
for sample in fd_movie:
    freqratio[sample] = float(fd_movie[sample]/ float(5+ fd_movie[sample] + 1.2736*fd_brown[sample]) )
    #adjust freqs in "brown" for difference in number of words.
    
from operator import itemgetter
sorted_freqratio = sorted(freqratio.items(), key = itemgetter(1), reverse = True)

for w, r in sorted_freqratio:
  print str(w) + ' ' + str(r)+' '+str(fd_movie[w])+' '+str(fd_brown[w])
```

```python
#divide between english words and non-english words (i.e. names, hopefully...)
from nltk.corpus import wordnet
from nltk.corpus import names

fname = open("freq_topic_names", "w")
fword = open("freq_topic_words", "w")

for w, r in sorted_freqratio:
    if not wordnet.synsets(w): #not a word
        fname.write(str(w) + ' ' + str(r)+'\n')
    elif w.title() in names.words():
        fname.write(str(w) + ' ' + str(r)+'\n')
    else:
        fword.write(str(w) + ' ' + str(r)+'\n')

fname.close()
fword.close()
```
